```python
import re
import import_ipynb
import lengthy_responses as long


def message_probability(user_message, recognised_words, single_response=False, required_words=[]):
    message_certainty = 0
    has_required_words = True

    # Counts number of words present in each message
    for word in user_message:
        if word in recognised_words:
            message_certainty += 1

    # find the percent of recognised words in a message by user
    percentage = float(message_certainty) / float(len(recognised_words))

    # Checks required words are in the string format or not
    for word in required_words:
        if word not in user_message:
            has_required_words = False
            break

    # Must either have the required words, or be a single response
    if has_required_words or single_response:
        return int(percentage * 100)
    else:
        return 0


def check_all_messages(message):
    highest_prob_list = {}

    # Simplifies response creation / adds it to the dict
    def responses(bot_response, list_of_words, single_response=False, required_words=[]):
        nonlocal highest_prob_list
        highest_prob_list[bot_response] = message_probability(message, list_of_words, single_response, required_words)

    # Responses -------------------------------------------------------------------------------------------------------
    responses('Hello!', ['hello', 'hi', 'hey','HY' 'sup', 'heyo'], single_response=True)
    responses('I\'m doing fine, and you?', ['how', 'are', 'you', 'doing'], required_words=['how'])
    responses('You\'re welcome!', ['thank', 'thanks'], single_response=True)
    responses('See you!', ['bye', 'goodbye'], single_response=True)
    responses('I am an adult', ['age', 'old'], single_response=True)
    responses('I am built by a professional developer Hasnain', ['w', 'goodbye'], single_response=True)
    responses('I am bot not a human!', ['married', 'single',"sex"], single_response=True)
    responses('Python!', ['language', 'Language you like',"programming language"], single_response=True)
    responses('Please you approprite Language!', ['Fuck', 'Sexy','bitch','Asshole'], single_response=True)
    responses('Thank you!', ['i', 'love', 'code', 'palace'], required_words=['code', 'palace'])

    # Longer responses
    responses(long.R_ADVICE, ['give', 'advice'], required_words=['advice'])
    responses(long.R_EATING, ['what', 'you', 'eat'], required_words=['you', 'eat'])

    best_match = max(highest_prob_list, key=highest_prob_list.get)
    return long.unknown() if highest_prob_list[best_match] < 1 else best_match


# Used to get the response
def get_responses(user_input):
    split_message = re.split(r'\s+|[,;?!.-]\s*', user_input.lower())
    responses = check_all_messages(split_message)
    return responses


# Testing the  system
while True:
    print('Bot: ' + get_responses(input('You: ')))
```
